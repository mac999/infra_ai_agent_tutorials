## Forward propagation with numpy

```python
import numpy as np 

# 가중치(W)와 편향(b)을 각 계층의 뉴런 수에 맞춰 랜덤 값으로 초기화
W1 = np.random.randn(4, 3) # 입력층(뉴런 3개)에서 은닉층1(뉴런 4개)로의 가중치 행렬 (4x3)
b1 = np.random.randn(4, 1) # 은닉층1의 각 뉴런에 대한 편향 벡터 (4x1)
W2 = np.random.randn(4, 4) # 은닉층1(뉴런 4개)에서 은닉층2(뉴런 4개)로의 가중치 행렬 (4x4)
b2 = np.random.randn(4, 1) # 은닉층2의 각 뉴런에 대한 편향 벡터 (4x1)
W3 = np.random.randn(1, 4) # 은닉층2(뉴런 4개)에서 출력층(뉴런 1개)으로의 가중치 행렬 (1x4)
b3 = np.random.randn(1, 1) # 출력층의 뉴런에 대한 편향 벡터 (1x1)

f = lambda x: 1.0/(1.0 + np.exp(-x)) # 활성화 함수로 시그모이드(sigmoid) 함수를 람다(lambda) 형식으로 정의
x = np.random.randn(3, 1)   # 3개의 원소를 가진 랜덤 입력 벡터 x (3x1 크기)를 생성
h1 = f(np.dot(W1, x) + b1)  # 첫 번째 은닉층의 가중치, 편향을 적용하고, 활성화(출력) 값을 계산
h2 = f(np.dot(W2, h1) + b2) # 두 번째 은닉층의 가중치, 편향을 적용하고, 활성화(출력) 값을 계산
out = np.dot(W3, h2) + b3   # 최종 출력 뉴런의 값을 계산
print(out)                  # 최종 출력 값을 출력
```
