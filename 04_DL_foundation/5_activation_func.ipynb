## Activation function

#### Sigmoid

```python
import torch, torch.nn as nn

activation_fn = nn.Sigmoid()   # Sigmoid 활성화 함수 인스턴스 생성
input_tensor = torch.tensor([-2.0, 0.0, 3.0])
output = activation_fn(input_tensor)   # Sigmoid 적용: 모든 값이 0과 1 사이로 압축

print(f"Input: {input_tensor}")
print(f"Sigmoid Output: {output}")
```

#### Tanh

```python
import torch, torch.nn as nn

activation_fn = nn.Tanh()   # Tanh 활성화 함수 인스턴스 생성
input_tensor = torch.tensor([-2.0, 0.0, 3.0])    # 가상의 입력값 (음수, 0, 양수)
output = activation_fn(input_tensor)    # Tanh 적용: 모든 값이 -1과 1 사이로 압축됨

print(f"Input: {input_tensor}")
print(f"Tanh Output: {output}") 
```

#### ReLU

```python
import torch.nn as nn

activation_fn = nn.ReLU()   # ReLU 활성화 함수 인스턴스 생성
input_tensor = torch.tensor([-2.0, 0.0, 3.0])

output = activation_fn(input_tensor)   # ReLU 적용: 음수는 0이 되고, 양수는 그대로 유지됨
print(f"Input: {input_tensor}")
print(f"ReLU Output: {output}")
```
