## Solution optimization

```python
import torch
import torch.nn as nn

model = nn.Linear(10, 1)    # 10개 입력을 1개 출력으로 바꾸는 선형 모델
x = torch.randn(1, 10)      # 임의의 입력 데이터
y = torch.randn(1, 1)       # 임의의 정답 데이터
criterion = nn.MSELoss()    # 평균 제곱 오차 손실

optimizer = torch.optim.SGD(model.parameters(), lr=0.01, momentum=0.9) # SGD 모멘텀 적용 옵티마이저. 학습률은 0.01
optimizer
```

```python
optimizer.zero_grad()   # 이전 기울기 값 초기화
outputs = model(x)      # 모델 예측
loss = criterion(outputs, y) # 손실 계산
loss.backward()         # 기울기 계산 (역전파)
optimizer.step()        # 모멘텀이 적용된 규칙으로 가중치를 업데이트합니다.

print(f'loss={loss.item()}')
```

```python
optimizer = torch.optim.RMSprop(model.parameters(), lr=0.01)  # RMSprop 최적화 설정. 각 파라미터 변화량을 자동 조절해 학습 안정성 향상
optimizer 
```

```python
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)  # Adam optimizer 적용. 학습률은 0.01
optimizer
```
