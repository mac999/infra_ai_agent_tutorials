## Tensor calculation

#### Define tensor

```python
import torch # PyTorch 라이브러리 가져오기

# 0차원 텐서 (스칼라)
scalar = torch.tensor(7)
print(scalar)

# 1차원 텐서 (벡터)
vector = torch.tensor([1.0, 2.0, 3.0])
print(vector)

# 2차원 텐서 (행렬)
matrix = torch.tensor([[1.0, 2.0], [3.0, 4.0]])
print(matrix)
```

#### Shape property

```python
print(scalar.shape) 
print(vector.shape)
print(matrix.shape)
```
